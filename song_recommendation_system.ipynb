```python
import numpy as np
import pandas as pd
import os 
import difflib
os.chdir("C:\\Users\\shoaib\\Downloads")
import Recommenders as Recommenders

#Load the Data - triplets_file
song_df_1 = pd.read_csv('triplets_file.csv')
print ("\nShape of song dataframe 1: ",song_df_1.shape)
print ("\nHead of song dataframe 1: \n",song_df_1.head())

#Load the Data - song_data
song_df_2 = pd.read_csv('song_data.csv')
print ("\nShape of song dataframe 2: ",song_df_2.shape)
print ("\nHead of song dataframe 2: \n",song_df_2.head())

#Combine two data frames and create one data Frame
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on='song_id', how='left')
print ("\nShape of consolidated data frame: ",song_df.shape)
print ("\nHead of consolidated data frame: \n",song_df.head())

#Display the length of each data Frame 
print("\nlength of dataframes of song dataframe 1 and song dataframe 2 respectively: ",len(song_df_1),"and", len(song_df_2))

#Length of the consolidated data frame 
print("\nLength of the consolidated data frame: ",len(song_df))

#Select only 50000 records to create a models to improve performance 
song_df = song_df.head(50000)
print("\nShape of song dataframe : ",song_df.shape)

#Creating new feature combining title and artist name
song_df['song'] = song_df['title']+' - '+song_df['artist_name']
print("\nHead of song dataframe: \n",song_df.head())

#Cumulative sum of listen count of the songs
song_grouped = song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()

#Sum of count – This is same as total number of records 
grouped_sum = song_grouped['listen_count'].sum()

#Display the percentage to identify most popular song 
song_grouped['percentage'] = (song_grouped['listen_count'] / grouped_sum ) * 100
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1])


#Import Popularity Recommender Model 
print("\nRecommendation based on popularity :")
pr = Recommenders.popularity_recommender_py()
pr.create(song_df, 'user_id', 'song')

#Display the top 10 popular songs- User 100 This is popularity rating same for all users 
display(pr.recommend(song_df['user_id'][100]))


#Item Similarity Recommendation
#Import Item Similarity Model 
ir = Recommenders.item_similarity_recommender_py()
ir.create(song_df, 'user_id', 'song')

#Item Similarity for User -5 
user_items = ir.get_user_items(song_df['user_id'][5])
# display user songs history
print("\nUser-5 songs history: \n")
for user_item in user_items:
    print(user_item)

#Item Similarity for User -100
user_items = ir.get_user_items(song_df['user_id'][100])
# display user songs history
print("\nUser-100 songs history: \n")
for user_item in user_items:
    print(user_item)

print("\nRecommendation based on item similarity :")
#Give song recommendation for that user-5
print("\nsong recommendation for user-5: ")
display(ir.recommend(song_df['user_id'][5]))

#Give song recommendation for that user-100
print("\nsong recommendation for user-100: ")
display(ir.recommend(song_df['user_id'][100]))

#Recommendation for another song
#Enter the song name to get another song recommendation
song_name = input('Enter your favourite song name : ')
list_of_all_titles = song_df['title'].tolist()
find_close_match = difflib.get_close_matches(song_name, list_of_all_titles)
close_match = find_close_match[0]
print("close match to your song name is: ",close_match)
songg=song_df[song_df.title == close_match]['song'].values[0]
print("\nNext song recommendation for you: ")
display(ir.get_similar_items([songg]))
```
